# Step 6: Create a spending flag for each user based on the average price across all their orders

```python
# Group data by "user_id" column, and create a mean/product aggregation and column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output
ords_prods_merge.head()
```

```python
# Create flags 
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts (dropna=False)
```

```python
ords_prods_merge.head(100)
```

# Step 7: Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column.

```python
# Group data by "user_id" column
ords_prods_merge['median_on_days'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Check output
ords_prods_merge.head(100)
```

```python
# Create flags 
ords_prods_merge.loc[ords_prods_merge['median_on_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_on_days'] > 10) & (ords_prods_merge['median_on_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_on_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check output
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
# Check null values
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

### Observation: the null values are from customers who have only ordered once.

```python
# Changing the null values to "Non-frequent customer"
ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
